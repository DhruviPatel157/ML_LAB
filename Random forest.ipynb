# Importing All required Modules and Models

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import seaborn as sns
```

# Loading dataset into variable

```python
digits = load_digits()
dir(digits)
```

# Importing train test split function and dividing data into train and test part

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state = 0)
```

# Importing RandomForestClassifier from sklearn and initializing object

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=1000)
```

# Fitting value of training and testing data and checking score

```python
model.fit(x_train, y_train)
print("SCORE : ", model.score(x_test, y_test))
y_pred = model.predict(x_test)
```

# metrics to see overall report of predicted value and test values

```python
from sklearn import metrics
print(metrics.classification_report(y_pred, y_test))
```

# Plotting confusion matrix using seaborn library

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
plt.figure(figsize =(10, 5))
mat = confusion_matrix(y_test, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
