# Importing All required Modules and Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

# Loading dataset into variable

```python
iris = load_iris()
```

```python
iris.feature_names
class_names = iris.target_names
```

# Importing train test split function and dividing data into train and test part

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state = 7)
```

# Logistic Regression Fitting data and score of model

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

# Plotting confusion matrix

```python
y_pred = lr.predict(x_test)
print(confusion_matrix(y_test, y_pred))
plot_confusion_matrix(lr, x_test, y_test, display_labels=class_names, cmap=plt.cm.Reds)
```

# Using cross_val_score with SVC

```python
svm =SVC()
svm.fit(x_train, y_train)
svm.score(x_test, y_test)
```

# Plotting confusion matrix

```python
y_pred = svm.predict(x_test)
print(confusion_matrix(y_test, y_pred))
plot_confusion_matrix(svm, x_test, y_test, display_labels=class_names, cmap=plt.cm.Reds)
```

# KNN Fitting data and score of model

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

# Plotting confusion matrix

```python
y_pred = knn.predict(x_test)
print(confusion_matrix(y_test, y_pred))
plot_confusion_matrix(knn, x_test, y_test, display_labels=class_names, cmap=plt.cm.Reds)
```
